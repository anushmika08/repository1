```python
from sklearn.preprocessing import LabelEncoder


le = LabelEncoder()

labelfit = le.fit(df['experience_level'])

df['experience_level'] = labelfit.transform(df['experience_level'])

df.head()
```

```python
from sklearn.preprocessing import LabelEncoder


le = LabelEncoder()

labelfit = le.fit(df['employment_type'])

df['employment_type'] = labelfit.transform(df['employment_type'])

df.head()
```

```python
from sklearn.preprocessing import LabelEncoder


le = LabelEncoder()

labelfit = le.fit(df['job_title'])

df['job_title'] = labelfit.transform(df['job_title'])

df.head()
```

```python
from sklearn.preprocessing import LabelEncoder


le = LabelEncoder()

labelfit = le.fit(df['salary_currency'])

df['salary_currency'] = labelfit.transform(df['salary_currency'])

df.head()
```

```python
from sklearn.preprocessing import LabelEncoder


le = LabelEncoder()

labelfit = le.fit(df['company_location'])

df['company_location'] = labelfit.transform(df['company_location'])

df.head()
```

```python
def convert_company_size(size):
    if size == 'S':
        return 0
    elif size == 'M':
        return 1
    elif size == 'L':
        return 2
    else:
        return 3
```

```python
df['company_size'] = df['company_size'].apply(convert_company_size)

df.head()
```

#Bar plot

```python
import matplotlib.pyplot as plt

import seaborn as sns


plt.figure(figsize=(10, 6))

sns.barplot(x='experience_level',y='salary_in_usd', data=df)

plt.title('Salary vs Experience Level')

plt.xlabel('Experience Level')


plt.ylabel('Salary in USD')


plt.show()
```

#Heat Map

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

#Line Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(10, 6))
sns.lineplot(x='experience_level', y='salary_in_usd', data=df)
plt.title('Salary vs Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary in usd')
plt.show()
```

# Box Plot


```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(10, 6))
sns.boxplot(x='company_size', y='salary_in_usd', data=df)
plt.title('Salary Distribution by Company Size')
plt.xlabel('Company Size')
plt.ylabel('Salary in USD')
plt.show()
```

# Violin Plot
